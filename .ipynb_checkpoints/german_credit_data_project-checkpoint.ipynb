```python
features = ["Age", "Sex", "Job", "Housing", "Saving accounts", "Checking account", "Credit amount", "Duration"]
```

```python
target = ["Risk"]
```

```python
df_model = df[features + target].copy()
```

```python
df_model.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
import joblib
```

```python
cat_col = df_model.select_dtypes(include = object).columns.drop("Risk")
```

```python
cat_col
```

```python
label_encoder_dict = {}
```

```python
for col in cat_col:
    la = LabelEncoder()
    df_model[col] = la.fit_transform(df_model[col])
    label_encoder_dict[col] = la
    joblib.dump(la, f"{col}_encoder.pkl")
```

```python
target
```

```python
le_target = LabelEncoder()
```

```python
df_model["Risk"] = le_target.fit_transform(df_model["Risk"])
```

```python
joblib.dump(le_target, "target_encoder.pkl")
```

```python
df_model.head(10)
```

```python
from sklearn.model_selection import train_test_split
```
